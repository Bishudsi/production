# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
# Load the libraries as required.
import pandas as pd
import pickle
import shap
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, PowerTransformer, OneHotEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import make_scorer, mean_squared_error
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fire_ds = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
fire_ds.info()

#fire_ds.head()
```

# Get X and Y

Create the features data frame and target data.

```python
fire_num_cols1 = ['coord_x', 'coord_y', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'wind', 'rh','rain']
fire_num_cols2 = ['wind', 'rain']
```

```python
fire_chr_cols = [ 'month' , 'day' ]
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
preproc1 = ColumnTransformer([
    ('num_scaled', StandardScaler(), fire_num_cols1),
    ('one_hot_enc', OneHotEncoder(handle_unknown='infrequent_if_exist'), fire_chr_cols)],
 remainder='passthrough')

preproc1
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
nonlinear_trans = Pipeline([
    ('standardizer', StandardScaler()),
    ('transform', PowerTransformer(method='yeo-johnson'))
])


preproc2 = ColumnTransformer([
    ('num_scaled', StandardScaler() , list(set(fire_num_cols1) - set(fire_num_cols2))),
    ('non_linear', nonlinear_trans , fire_num_cols2),
    ('one_hot_enc', OneHotEncoder(handle_unknown='infrequent_if_exist'), fire_chr_cols)],
 remainder='passthrough')

preproc2
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.
+ Add a step labelled `regressor` and assign a regression model to it. 

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
# Pipeline A = preproc1 + baseline
PipelineA = Pipeline([
    ('preprocessing', preproc1),
    ('regressor', KNeighborsRegressor(n_neighbors=5))
])
```

```python
# Pipeline B = preproc2 + baseline
PipelineB = Pipeline([
    ('preprocessing', preproc2),
    ('regressor', KNeighborsRegressor(n_neighbors=5))
])
```

```python
# Pipeline C = preproc1 + advanced model
PipelineC = Pipeline([
    ('preprocessing', preproc1),
    #('regressor', DecisionTreeRegressor(random_state=42))
    ('regressor', RandomForestRegressor(random_state=42))
])
```

```python
# Pipeline D = preproc2 + advanced model
PipelineD = Pipeline([
    ('preprocessing', preproc2),
    ('regressor', RandomForestRegressor (random_state=42))
])
    
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
X = fire_ds.drop(columns = 'area')
Y = fire_ds['area']

scoring = {
    'r2': 'r2',
    'rmse': make_scorer(mean_squared_error, squared=False)
}

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
param_gridPA = {
    'regressor__n_neighbors' : [ 2, 5, 6, 9 ] 
 }

grid_shcv_PA = GridSearchCV(
    estimator=PipelineA, 
    param_grid=param_gridPA, 
    scoring=scoring, 
    cv = 5,
    refit = 'rmse')
grid_shcv_PA.fit(X_train, Y_train)
grid_shcv_PA.best_params_, grid_shcv_PA.best_score_
```

```python
param_gridPB = {
    'regressor__n_neighbors' : [ 2, 5, 6, 9 ] 
 }

grid_shcv_PB = GridSearchCV(
    estimator=PipelineB, 
    param_grid=param_gridPB, 
    scoring=scoring, 
    cv = 5,
    refit = 'rmse')

grid_shcv_PB.fit(X_train, Y_train)
grid_shcv_PB.best_params_, grid_shcv_PB.best_score_
```

```python
param_gridPC = {
    'regressor__n_estimators' : [ 25, 50] 
 }

grid_shcv_PC = GridSearchCV(
    estimator=PipelineC, 
    param_grid=param_gridPC, 
    scoring=scoring, 
    cv = 5,
    refit = 'rmse')

grid_shcv_PC.fit(X_train, Y_train)
grid_shcv_PC.best_params_, grid_shcv_PC.best_score_
```

```python
param_gridPD = {
    'regressor__n_estimators' : [ 25, 50] 
 }

grid_shcv_PD = GridSearchCV(
    estimator=PipelineD, 
    param_grid=param_gridPD, 
    scoring=scoring, 
    cv = 5,
    refit = 'rmse')

grid_shcv_PD.fit(X_train, Y_train)
grid_shcv_PD.best_params_, grid_shcv_PD.best_score_

#preds = grid_shcv_PD.best_estimator_.predict(X_test)
#preds
```
